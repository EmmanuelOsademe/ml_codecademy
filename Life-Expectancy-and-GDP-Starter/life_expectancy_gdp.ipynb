```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("all_data.csv")
print(df.head())
print(df.Country.unique())
print(df.Year.unique())
```

```python
df.info()
```

```python
df = df.rename(columns={"Life expectancy at birth (years)": "Expectancy"})
columns = df.columns
# print(columns)

mean_expectancy = df.groupby('Country').Expectancy.mean()
print(mean_expectancy)
mean_gdp = df.groupby("Country").GDP.mean()

x_values = range(len(mean_expectancy))
x_labels = ["Chile", "China", "Germany", "Mexico", "USA", "Zimbabwe"]
ax = plt.subplot()
plt.bar(x_values, mean_expectancy)
ax.set_xticks(x_values)
ax.set_xticklabels(x_labels)
plt.xlabel("Country")
plt.ylabel("Time-averaged Life Expantancy")
plt.show()
plt.clf()
```

```python
mean_gdp = df.groupby("Country").GDP.mean()/1000000000
x_values = range(len(mean_gdp))
x_labels = ["Chile", "China", "Germany", "Mexico", "USA", "Zimbabwe"]
ax = plt.subplot()
plt.bar(x_values, mean_gdp)
ax.set_xticks(x_values)
ax.set_xticklabels(x_labels)
plt.xlabel("Country")
plt.ylabel("Time-averaged GDP")
plt.show()
plt.clf()
```

```python
countries = df.Country.unique()

def generate_lineplot_by_countries(variable):
    for country in countries:
        data = df[df.Country == country]
        plt.plot(data.Year, data[variable])
        plt.show()
        plt.clf()
```
